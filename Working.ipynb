```python
img_shape = 224
filename = "hand1_a_bot_seg_2_cropped.jpeg"
# Read in target file (an image)
img = tf.io.read_file(filename)
# Decode the read file into a tensor & ensure 3 colour channels
# (our model is trained on images with 3 colour channels and sometimes images have 4 colour channels)
img = tf.image.decode_image(img, channels=3)
# Resize the image (to the same size our model was trained on)
img = tf.image.resize(img, size = [img_shape, img_shape])
# Rescale the image (get all values between 0 and 1)
img = img/255
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def pred_and_plot(model,class_names):
  # Import the target image and preprocess it
  # Make a prediction
  pred = model.predict(tf.expand_dims(img, axis=0))

  # Get the predicted class
  pred_array = np.array(pred)
  print(pred_array)
  pred_class = class_names[np.argmax(pred_array)]

  # Plot the image and predicted class
  plt.imshow(img)
  plt.title(f"Prediction: {pred_class}")
  plt.axis(False);
```

```python
class_names = ['a','b','c','d','e','f','g','h']
```

```python
pred_and_plot(model,class_names)
```

```python
import tkinter as tk
from PIL import ImageTk, Image
from tkinter import filedialog

# Create the main window
window = tk.Tk()
window.geometry('400x400')
window.title("ASL Translator")
# Function to open and display an image
def open_image():
    # Open a file dialog to browse system files
    global file_path
    file_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.jpg;*.jpeg;*.png")])

    # Check if a file was selected
    if file_path:
        # Open the image using PIL
        global image 
        image = Image.open(file_path)

        # Resize the image to fit the window
        width, height = image.size
        if width > 800 or height > 600:
            ratio = min(800 / width, 600 / height)
            width = int(width * ratio)
            height = int(height * ratio)
            image = image.resize((width, height), Image.ANTIALIAS)

        # Create a Tkinter-compatible photo image
        photo = ImageTk.PhotoImage(image)

        # Create a label to display the image
        label = tk.Label(window, image=photo)
        label.image = photo
        label.pack()
        

#Function to be executed when the button is clicked
def custom_function():
    img_shape = 224
    # Read in target file (an image)
    img = tf.io.read_file(file_path)
    # Decode the read file into a tensor & ensure 3 colour channels
    # (our model is trained on images with 3 colour channels and sometimes images have 4 colour channels)
    img = tf.image.decode_image(img, channels=3)
    # Resize the image (to the same size our model was trained on)
    img = tf.image.resize(img, size = [img_shape, img_shape])
    # Rescale the image (get all values between 0 and 1)
    img = img/255
    pred = model.predict(tf.expand_dims(img, axis=0))

    # Get the predicted class
    pred_array = np.array(pred)
    pred_class = class_names[np.argmax(pred_array)]

    # Plot the image and predicted class
    plt.imshow(img)
    plt.title(f"Prediction: {pred_class}")
    plt.axis(False);

# Create a button to open an image and execute the custom function
button = tk.Button(window, text="Open Image", command=open_image)
button.pack()

# Create a button to execute the custom function
custom_button = tk.Button(window, text="Translate", command=custom_function)
custom_button.pack()

#Text Label
'''label = tk.Label(window, text="hello",font=20)
label.pack()'''

# Run the main window's event loop
window.mainloop()
```
